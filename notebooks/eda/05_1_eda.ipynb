```python
import pathlib as pl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
root_path = pl.Path.cwd().parent.parent
data_path = root_path / "data" / "intermed"
```

```python
data_review = pd.read_csv(data_path/"googleplaystore_user_reviews_clean.csv")
data_apps = pd.read_csv(data_path/"googleplaystore_clean.csv")
```

```python
merged_data = pd.merge(left=data_apps,right=data_review,how="inner",on=["App"])
```

```python
polarizacao_df = merged_data.groupby('Genres').agg({
    'Sentiment_Polarity': 'mean', 
    'Rating':'mean',
    'Translated_Review': 'count'
}).dropna() 

polarizacao_df.columns = ['Media_Sentimento', 'Media_Avaliacao', 'Total_Reviews']

polarizacao_df = polarizacao_df[polarizacao_df['Total_Reviews'] >= 20]
polarizacao_df['Gap'] = abs(polarizacao_df['Media_Sentimento'] - polarizacao_df['Media_Avaliacao'])

apps_polarizados = polarizacao_df.sort_values(by='Gap', ascending=False)

print("=== OS APLICATIVOS MAIS POLARIZADOS (AME OU ODEIE) ===")
print(apps_polarizados.head(10))
```
