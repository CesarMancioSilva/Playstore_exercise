```python
import pandas as pd
import seaborn as sns
import pathlib as pl
import numpy as np
import matplotlib.pyplot as plt
```

```python
root_path = pl.Path.cwd().parent.parent
data_path = root_path / "data" / "intermed"
```

```python
data = pd.read_csv(data_path/"googleplaystore_clean.csv")
```

```python
data["Last Updated"] = pd.to_datetime(data["Last Updated"],errors="coerce")
```

```python
data["Last Updated"].describe()
```

```python

resumo_generos = data.groupby('Genres')['Days_Since_Update'].mean().reset_index()

top_15_recentes = resumo_generos.sort_values(by='Days_Since_Update',ascending=False).head(20)

plt.figure(figsize=(12, 7))

sns.barplot(data=top_15_recentes, x='Days_Since_Update', y='Genres', palette='viridis')

plt.title('Top 20 Gêneros com Atualizações Mais Atrasadas', fontsize=16, pad=10)
plt.xlabel('Média de Dias Desde a Última Atualização (Menor = Mais Recente)', fontsize=12)
plt.ylabel('Gêneros de Aplicativos', fontsize=12)

plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
